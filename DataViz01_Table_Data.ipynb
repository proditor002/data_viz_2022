```python
data_file_path = os.path.join(path,'confirmed-cases.xlsx')     # ไฟล์ที่เราต้องการ  
dataHW_covid = pd.read_excel(data_file_path)      # .read_excel อ่านไฟล์ที่เป็นexcel
```

```python
dataHW_covid #แสดงข้อมูลตาราง dataHW_covid
```

## เลือกมาใช้เฉพาะข้อมูลที่ต้องการ

```python
data_HW = dataHW_covid[['sex','age','risk']] #เลือกเอามาใช้เฉพาะข้อมูลที่ต้องการ
data_HW
```

### ตรวจสอบค่า Missing

```python
data_HW.isnull().any() #ตรวจสอบค่า Missing ในตารางข้อมูลdata_HWcovid
```

```python
data_risk = data_HW.dropna() #แก้ Missing ด้วยการ Dropna
data_risk    #จากข้อมูล 816989 แถว 3 คอลัมน์ หลัง dropna เหลือ 712916 แถว 3 คอลัมน์
```

```python
data_risk.isnull().any() #ตรวจสอบค่า Missing อีกครั้ง
```

```python
set(data_risk['risk'])
```

```python
HW1 = data_risk[data_risk['risk']=='สนามมวย'] #เลือกเอาเฉพาะสนามมวย
HW1
```

```python
HW2 = data_risk[data_risk['risk']=='สถานบันเทิง'] #เลือกเอาเฉพาะสถานบันเทิง
HW2
```

```python
HW3 = data_risk[data_risk['risk']=='บุคลากรด้านการแพทย์และสาธารณสุข'] #เลือกเอาเฉพาะบุคลากรด้านการแพทย์และสาธารณสุข
HW3
```

### รวมตาราง

```python
data = pd.concat([HW1,HW2,HW3]) #รวมตาราง
data
```

```python
data_women = data.replace('หญิง',1) #กำหนดเพศหญิงเป็น1
data_women
```

```python
data_men  = data_women.replace('ชาย',50) #กำหนดเพศชายเป็น50
data_men
```

```python
data_t = data_men[['age','risk']] #เลือกเอาเฉพาะข้อมูลที่จะทำกราฟ
data_t
```

```python
set(data_men['risk'])
```
