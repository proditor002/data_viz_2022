#จัดการข้อมูลใหม่ให้อยู่ใน Scale ของจังหวัด

###จำนวนผู้ติดเชื้อรวม แบบรายจังหวัด

```python
covidpro = data_covidall.groupby('province_of_onset').size().reset_index(name='จำนวนผู้ติดเชื้อ')
covidpro
```

###แบ่งช่วงอายุเป็น 3 ช่วง และนับรวมแบบรายจังหวัด

```python
age1= data_covidall[data_covidall['age']<=14].groupby('province_of_onset').size().reset_index(name='อายุน้อยกว่า 15 ปี') # เลือกช่วงอายุที่น้อยกว่า 15 ปี และตั้งชื่อคอลัมน์ว่า อายุน้อยกว่า 15 ปี
age1
```

```python
age2= data_covidall[(data_covidall['age']>=15)&(data_covidall['age']<=59)].groupby('province_of_onset').size().reset_index(name='อายุ 15-59 ปี') # เลือกช่วงอายุ 15-59 ปี และตั้งชื่อคอลัมน์ว่า อายุ 15-59 ปี
age2
```

```python
age3= data_covidall[data_covidall['age']>=60].groupby('province_of_onset').size().reset_index(name='อายุ 60 ปีขึ้นไป') # เลือกช่วงอายุที่ 60 ปีขึ้นไป และตั้งชื่อคอลัมน์ว่า อายุ 60 ปีขึ้นไป
age3
```

####รวมตารางของอายุ

```python
agemerge = age1.merge(age2,how='left',left_on='province_of_onset',right_on='province_of_onset') # เชื่อมตารางของอายุ อายุน้อยกว่า 15 ปี และ	อายุ 15-59 ปี เข้าด้วยกัน โดยใช้คอลัมน์ 'province_of_onset'เป็นตัวเชื่อม
agemerge
```

```python
agemergeall = agemerge.merge(age3,how='left',left_on='province_of_onset',right_on='province_of_onset')  # เชื่อมตารางของ อายุ 60 ปีขึ้นไป เข้ากับตารางก่อนหน้า โดยใช้คอลัมน์ 'province_of_onset'เป็นตัวเชื่อม
agemergeall
```

###แบ่งเพศออกเป็น 3 แบบ และนับรวมแบบรายจังหวัด

```python
female= data_covidall[data_covidall['sex']=='หญิง'].groupby('province_of_onset').size().reset_index(name='หญิง') #เลือกเฉพาะเพศหญิง  และใช้.groupby เพื่อจัดกลุ่มตามจังหวัด ใช้.size เพื่อนับขนาด และตั้งชื่อคอลัมน์ว่า 'หญิง' 
female
```

```python
male= data_covidall[data_covidall['sex']=='ชาย'].groupby('province_of_onset').size().reset_index(name='ชาย') #เลือกเฉพาะเพศชาย  และใช้.groupby เพื่อจัดกลุ่มตามจังหวัด ใช้.size เพื่อนับขนาด และตั้งชื่อคอลัมน์ว่า 'ชาย'
male
```

```python
nogender= data_covidall[data_covidall['sex']=='ไม่ระบุ'].groupby('province_of_onset').size().reset_index(name='ไม่ระบุ') #เลือกเฉพาะไม่ระบุ  และใช้.groupby เพื่อจัดกลุ่มตามจังหวัด ใช้.size เพื่อนับขนาด และตั้งชื่อคอลัมน์ว่า 'ไม่ระบุ'
nogender
```

####รวมตารางเพศ

```python
data1 = female.merge(male,how='left',left_on='province_of_onset',right_on='province_of_onset') #รวมตารางเพศชาย กับ เพศหญิง
data1
```

```python
data2 = data1.merge(nogender,how='left',left_on='province_of_onset',right_on='province_of_onset') # รวมตารางไม่ระบุเพศ เข้ากับ ตารางก่อนหน้า
data2
```
