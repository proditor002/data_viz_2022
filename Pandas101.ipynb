#### HW (ให้ใช้เฉพาะที่อาจารย์สอนไปแล้วในวิชานี้)

1.   คำนวณ อายุเฉลี่ย ของผู้ชาย ในข้อมูลทั้งหมด
2.   คำนวณ อายุเฉลี่ย ของผู้ชาย ในผู้ป่วยในจังหวัดขอนแก่น
3. หาจำนวนผู้ป่วยที่เป็นคน "**ไม่ใช้คนไทย**"

````
data = data_covid[['sex','age','province_of_onset']] # เลือกมาเฉพาะ column ที่ต้องการใช้
````

````
data # print ออกมาดู
````

````
male = data[data['sex']=='ชาย'] # เลือกมาเฉพาะค่าที่เป็นผู้ชาย
male['age'] # หาอายุของผู้ชาย
````

````
#1.คำนวณ อายุเฉลี่ย ของผู้ชาย ในข้อมูลทั้งหมด
sum = 0 # sum มารับโดยให้เป็น 0 เนื่องจากจะให้มารับค่าทั้งหมดที่จะนำมาบวกกัน
N = 0 # N โดยนับจำนวนคนทั้งหมด เริ่มที่ 0
for a in male['age']: # นำค่าอายุของผู้ชายเข้ามาเพื่อวนดูค่าอายุ
    
    if a > 0: # ถ้า a มากกว่า 0 แสดงว่าเป็นอายุของคน จะนำมานับ
        sum += a # sum = sum + a # คำนวณหาอายุของผู้ชาย ซึ่งถ้า a นั้นเป็นอายุคนก็จะถูกนับมาบวกกันเรื่อย ๆ
        N += 1 # คำนวณหาจำนวนผู้ชายทั้งหมดโดยเพิ่มทีละ 1

print(f'อายุเฉลี่ย ของ ผู้ป่วยชาย {sum/N}') #print ค่าเฉลี่ยของผู้ป่วยชาย โดยนำ sum = อายุรวมทั้งหมดของผู้ป่วยชาย หารด้วย N = จำนวนผู้ป่วยชายทั้งหมด
````

````
#2.คำนวณ อายุเฉลี่ย ของผู้ชาย ในผู้ป่วยในจังหวัดขอนแก่น
male_KK = male[male['province_of_onset']=='ขอนแก่น'] # เลือกค่าที่เป็นผู้ชายที่อยู่ในจังหวัดขอนแก่น

sum = 0 # sum มารับโดยให้เป็น 0 เนื่องจากจะให้มารับค่าทั้งหมดที่จะนำมาบวกกัน
N = 0 # N โดยนับจำนวนคนทั้งหมด เริ่มที่ 0
for a in male_KK['age']: # นำค่าอายุของผู้ชายเข้ามาเพื่อวนดูค่าอายุ 
    
    if a > 0: # ถ้า a มากกว่า 0 แสดงว่าเป็นอายุของคน จะนำมานับ
        sum += a # sum = sum + a # sum = sum + a # คำนวณหาอายุของผู้ชาย ซึ่งถ้า a นั้นเป็นอายุคนก็จะถูกนับมาบวกกันเรื่อย ๆ
        N += 1 # คำนวณหาจำนวนผู้ชายทั้งหมดโดยเพิ่มทีละ 1

print(f'อายุเฉลี่ย ของ ผู้ป่วยชาย ในจังหวัดขอนแก่น {sum/N}') #print ค่าเฉลี่ยของผู้ป่วยชายในขอนแก่น โดยนำ sum = อายุรวมของผู้ป่วยชายในขอนแก่น หารด้วย N = จำนวนผู้ป่วยชายทั้งหมด
````

## Quiz ตัดตารางออกมาเฉพาะปี 2021 ```announce_date``` ในปี 2021 

**Hint**
*   วนลูปหา index ของปี 2021
*   ตัดตารางมาเฉพาะ ปี 2021



````
TF=list()
for each_row in data_covid.iterrows():
    if each_row[1]['announce_date'].split('/')[2] == '2021':
        TF.append(True)
    else:
        TF.append(False)

data_covid[TF].head()
````

### HW กลุ่ม 2 || เลือกข้อมูลมาเฉพาะวัน announce_date เป็นวันที่ 1-20

````
AD = list() # กำหนดให้ AD คือ list ว่าง เพื่อรับค่าที่จะใส่เข้ามา
for each_row in data_covid.iterrows(): # วนดูค่าในแต่ละแถวในตาราง data_covid
    if each_row[1]['announce_date'].split('/')[1] <= '20': # กำหนดเงื่อนไข โดยดูเลือกแถวที่ 1 ในตาราง announce_date โดยที่แยกตามเครื่อง / โดยเลือกดูค่าที่น้อยกว่าหรือเท่ากับ 20
        AD.append(True) # ถ้าตรงตามเงื่อนไข จะถูกเพิ่มเข้าไปใน list AD
    else: # โดยที่
        AD.append(False) # แต่ถ้าวันไม่ตรงตามเงื่อนไข จะไม่เอามาเก็บใน list AD

data_covid[AD] # print ค่าใน list ออกมาดู
````
