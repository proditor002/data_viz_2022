# ข้อมูลชุดที่ 1

```python
data1 = os.path.join(path,'hospital_list.xlsx') #แล้วก็ชี้ไปที่ไฟล์
print(data1)
```

```python
data2 = pd.read_excel(data1)  #กำหนดตัวแปร data2 เป็นข้อมูล ไฟล์ hospital_list.xlsx
data2  #เรียกดูตารางของข้อมูล ไฟล์ hospital_list.xlsx
```

## ตัดเอาเฉพาะคอลัมน์ที่ต้องการ

```python
data3= data2[['hcode','name','zone_name','province','amphur','tambon','name5']]
data3
```

## ตรวจสอบค่า missing

```python
data3.isnull().any()
```

## แทน missing ด้วยค่าที่เหมาะสม 

โดยคำสั่ง fillna() ของตาราง data4

```python
data4= data3.fillna(value={'name':'ไม่ทราบ','zone_name':'ไม่ทราบ','amphur':'ไม่ทราบ','tambon':'ไม่ทราบ','name5':'ไม่ทราบ'}) #แทน missing ด้วยค่าที่เหมาะสม จากตาราง data3
data4 # แสดงตาราง data4
```

##ตรวจสอบว่ายังมีค่า Missing หรือไม่

```python
data4.isnull().any()
```

# ข้อมูลชุดที่ 2

```python
data5 = os.path.join(path,'alldb.csv') #แล้วก็ชี้ไปที่ไฟล์
print(data5)
```

```python
data6 = pd.read_csv(data5)  #กำหนดตัวแปร data6 เป็นข้อมูล ไฟล์ alldb.csv
data6  #เรียกดูตารางของข้อมูล ไฟล์ alldb.csv
```

## ตัดเอาเฉพาะคอลัมน์ที่ต้องการ

```python
data7= data6[['user','RowKey','date','phone','site','time','urine','ov','PartitionKey']]
data7
```

## ตรวจสอบว่าข้อมูลมีค่า Missing หรือไม่

```python
data7.isnull().any()  #ดูค่า Missing รายคอลัมน์ของตาราง data6
#เนื่องจากมี True แสดงว่ามีค่าที่เป็น Missing จะทำการลบค่า Missing ต่อไป
```

## เปลี่ยน Type Data จากข้อมูลที่เป็น string ให้เป็นค่าว่าง

```python
data7['user'] = pd.to_numeric(data7['user'], errors='coerce')
data7['site'] = pd.to_numeric(data7['site'], errors='coerce')
data7['ov'] = pd.to_numeric(data7['ov'], errors='coerce')
data7['PartitionKey'] = pd.to_numeric(data7['PartitionKey'], errors='coerce')
data7
```

## เปลี่ยน Type Data จากข้อมูลที่เป็น numeric ให้เป็นค่าว่าง

```python
data7['phone']=data7['phone'].astype('string')
data7['RowKey']=data7['RowKey'].astype('string')
data7['urine']=data7['urine'].astype('string')
data7
```

## แทน missing ด้วยค่าที่เหมาะสม 

โดยคำสั่ง fillna() ของตาราง data8

```python
data8= data7.fillna(value={'user':'00000','RowKey':'ไม่ระบุ','date':'0000-00-00','phone':'ไม่ระบุ','site':'00000','time':'00:00:00','urine':'ไม่ระบุ','PartitionKey':'00000'}) #แทน missing ด้วยค่าที่เหมาะสม จากตาราง data7
data8 # แสดงตาราง data8
```

```python
data9 = data8.dropna() #ตัดข้อมูลที่เป็นค่าว่างของ ov ออก เพราะไม่สามารถแทนค่า missing ด้วยค่าอื่นได้ เนื่องจากต้องการข้อมูลที่เป็นตัวเลข 0-4 เท่านั้น
data9
```

##ตรวจสอบว่ายังมีค่า Missing หรือไม่

```python
data9.isnull().any()
```
